Para a침adir markers en un mapa, tendremos que "a침adir" elementos a dicho mapa, por eso, vamos a a침adirlos al mapa anterior llamado map_1. 

```python
#Marker va en may칰scula
parque= Marker(location=[inicial_lat, inicial_long], tooltip='Parque de Dobby')
parque.add_to(map_1) #lo a침adimos como una capa al mapa ya crado
casa= Marker(location=[40.40231152266822, -3.6722349888902435],tooltip='Casa')
casa.add_to(map_1) #lo a침adimos como una capa al mapa ya crado

map_1
```

Para modificar los iconos 쯗e d칩nde los sacamos?
- https://fontawesome.com/v4.7.0/icons/

```python
#prefix = fa lo ponemos cuando usemos los iconos de fontawesome.
icono= Icon(
            color = 'blue',
            prefix= 'fa',
            icon= 'paw',
            icon_color= 'black'
            )

loc = {"location": [lavapi_lat, lavapi_long], "tooltip": "Casa de Dobby"}
```

```python
#guardamos el marker en una variable
marker_cueva = Marker(**loc, icon = icono )
```

```python
#a침adimos el marker al mapa original donde ya ten칤amos el parque de Dobby guardado
marker_cueva.add_to(map_1)
map_1
```

```python
icono_iberswim= Icon(
            color = 'blue',
            prefix= 'fa',
            icon= 'swimmer',
            icon_color= 'white'
            )

iberswim_coor= [40.34577660801068, -3.699893886705866]
iberswim = {"location": iberswim_coor, "tooltip": "IBERSWIM"}
marker_iberswim= Marker(**iberswim, icon = icono_iberswim )
marker_iberswim.add_to(map_1)
```
